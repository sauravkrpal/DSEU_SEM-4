```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Encode the categorical 'weather' column into numerical values
label_encoder = LabelEncoder()
weather_data['weather_encoded'] = label_encoder.fit_transform(weather_data['weather'])

# Define features (precipitation, temp_max, temp_min, wind) and target (weather_encoded)
features = weather_data[['precipitation', 'temp_max', 'temp_min', 'wind']]
target = weather_data['weather_encoded']

# Split the data into training and testing sets (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Check the shape of the training data
X_train.shape, X_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Initialize the Random Forest classifier
rf_model = RandomForestClassifier(random_state=42)

# Train the model on the training data
rf_model.fit(X_train, y_train)

# Make predictions on the test data
y_pred = rf_model.predict(X_test)

# Evaluate the model's accuracy
accuracy = accuracy_score(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred, target_names=label_encoder.classes_)

print(accuracy)
print(classification_rep)
```

```python
import numpy as np

# Function to predict weather based on input features
def predict_weather(precipitation, temp_max, temp_min, wind):
    # Create a NumPy array for the input data
    input_data = np.array([[precipitation, temp_max, temp_min, wind]])

    # Predict the weather using the trained Random Forest model
    prediction_encoded = rf_model.predict(input_data)[0]

    # Decode the numeric prediction back to the original weather label
    predicted_weather = label_encoder.inverse_transform([prediction_encoded])[0]

    return predicted_weather

# Example usage: Provide precipitation, temp_max, temp_min, and wind values for prediction
precipitation = 5.0
temp_max = 15.0
temp_min = 8.0
wind = 3.0

predicted_weather = predict_weather(precipitation, temp_max, temp_min, wind)
print(f"Predicted Weather: {predicted_weather}")
```
