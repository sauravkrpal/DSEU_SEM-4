# Q2: Randomly selecting 5 data points from the dataset

```python
import numpy as np

# Randomly selecting 5 data points from the dataset
random_data_points = df.sample(5, random_state=42)

random_data_points
```

# Q3: Calculating Z-score for the 'Close' column in these data points

```python
# Calculate Z-score for the 'Close' column in these data points
mean_close = df['Close'].mean()
std_close = df['Close'].std()

# Z-score calculation
random_data_points['Z-Score Close'] = (random_data_points['Close'] - mean_close) / std_close

# Display the randomly selected data points and their Z-scores
random_data_points[['Date', 'Close', 'Z-Score Close']]
```
